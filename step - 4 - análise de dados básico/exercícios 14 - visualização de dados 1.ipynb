## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
dfc = data[["cut", "price"]].groupby("cut").mean().reset_index()

diamonds = sns.barplot(data=dfc,x=dfc.index, y="price", palette="pastel")
diamonds.set(title='Valor do diamante por tipo de corte', xlabel='Tipo de corte', ylabel='Valor do diamante (USD)');
```

**Insight do gráfico 1**:

o valor médio mais alto do corte premium e fair provavelmente são resultados de sertem os mais vendios mesmo não sneo tão caro quanto o ideal

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2
dfc = data.query("month == 'Dec'")

with sns.axes_style('whitegrid'):
  flights = sns.lineplot(data=dfc, x="year", y="passengers")
  flights.set(title='Número de passageiros em dezembro por ano', xlabel='Ano', ylabel='Número de passageiros');
```

**Insight do gráfico 2**:

pode-se observar que ao passar dos anos existe um aumento conxistente nas viagens em dezembro

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3
dfc = data.query("1949 <= year <= 1959")





with sns.axes_style('whitegrid'):
  flights = sns.lineplot(data=dfc, x="month", y="passengers", hue="year",palette="dark")
  flights.set(title='Número de passageiros por mês entre 1949 e 1959', xlabel='Mês', ylabel='Número de passageiros');
  flights.legend(title='Ano',prop={"size": 6.5, "weight": "bold"})
```

**Insight do gráfico 3**:

o numero de passageiros creçe de fevereiro a agosto tendo uma queda seguida de uma aumento e agosto até novembro em passageiros
