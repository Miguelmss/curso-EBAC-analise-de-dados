## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
with sns.axes_style("whitegrid"):
  diamonds = sns.boxplot(data=data["price"])
  diamonds.set_title("Preços de Diamantes")
```

**Insight do gráfico 1**:


os outliers do data frame diamonds se localizam acima de 10000 dolares, ou seja existe alguns diamentes que saõ quantitativamente mais caro que os demais.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

with sns.axes_style("whitegrid"):
  diamonds = sns.scatterplot(data=data, x="carat", y="price", hue="clarity")
  diamonds.set_title("Preços de Diamantes")
  diamonds.figure.set_size_inches(w=20/1.5, h=10/1.5)
```

**Insight do gráfico 2**:

pode-se analisar como boa parte dos diamantes se localizam entre 1 a 3 carat o
valor dentro deste intervalo vai de 2500 a mais de 1750 a transparencia não parece influenciar muito no começo sendo possivel ver uma mistura dela no começo com o aumentor do preço e valor fica preominante a transparencia sl1, outro informação seria que os diamentes de transparencia l1 seria, os mais pesao de forma média.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = sns.histplot(data=data, x="weight", kde=True)
cars.set_title("Distribuição do peso de carros")
```

**Insight do gráfico 3**:

pode-se observar que existe uma grande concentração de carros 2000 a 2500 kg que se da provavelmente por ser o pesso médio e carros populares que não possuem tanta potencia porem fazem uma otima kilometragem por litro graças ao seu peso e desing.
