## **Tarefa 2: Função para extração de coluna de arquivo csv** ##

1. A função deve ter como parâmetros nome_arquivo e indice_coluna.
1. Siga os passos apresentados pelas linhas de comentários no editor de código.

`````python
import os

def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int):
	coluna = []
	with open(file='../../data/' + nome_arquivo, mode="r",encoding='utf8') as arquivo:
		linha = arquivo.readline()# lê o cabeçalho
		linha = arquivo.readline()# lê a primeira linha
		while linha:
			linha_separada=linha.split(sep=",")
			colunas=linha_separada[indice_coluna]
			coluna.append(colunas)
			linha = arquivo.readline()
		
	# leia o arquivo com o comando 'with' utilizando o parametro 'nome_arquivo'
	# O caminho para o arquivo deve começar com '../../data/'
	# extraia a coluna do arquivo utilizando o parametro 'indice_coluna'

	return coluna

valor_manutencao = extrai_coluna_csv(nome_arquivo='carros.csv', indice_coluna=2)
print(valor_manutencao)
`````

## **Tarefa 3: Função para extração de coluna de arquivo csv e conversão de dados** ##

1. Complete a função para extração os valores de valor_venda e que retorne uma lista com os valores.
1. A função deve ter como parâmetros nome_arquivo, indice_coluna, tipo_dado.
1. O parâmetro tipo_dado deve aceitar como argumento os valores "str" ou "int".
1. Siga os passos apresentados pelas linhas de comentários no editor de código.

`````python
import os

def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int, tipo_dado: str):
	coluna = []
	with open(file='../../data/' + nome_arquivo, mode="r",encoding='utf8') as arquivo:
		linha = arquivo.readline()# lê o cabeçalho
		linha = arquivo.readline()# lê a primeira linha
		while linha:
			linha_separada=linha.split(sep=",")
			colunas=linha_separada[indice_coluna]
			if tipo_dado == "str":
				colunas=str(colunas)
			elif tipo_dado == "int":
				colunas=int(colunas)
			else:
				print("tipo de daddo incompativel")
			coluna.append(colunas)
			linha = arquivo.readline()

	# leia o arquivo com o comando 'with' utilizando o parametro 'nome_arquivo'.
    # O caminho para o arquivo deve começar com '../../data/'
	# extraia a coluna do arquivo utilizando o parametro 'indice_coluna'
	# use a estrutura de decisão if/elif/else para fazer a conversão do tipo de dados utilizando o parametro 'tipo_dado'

	return coluna

valor_manutencao = extrai_coluna_csv(nome_arquivo='carros.csv', indice_coluna=2, tipo_dado='str')
print(valor_manutencao)
`````
